```python
#Nessa aula, vamos aprender como criar estruturas
# condicionais aninhadas, usando os comandos
# if.. elif.. else em programas Python.

nome = str(input('Digite seu nome: ')).strip().capitalize()
if nome == 'Meirielly':
    print('Que nome bonito!')
elif nome == 'Paulo' or nome == 'José' or nome == 'Maria':
    print('Seu nome é bem popular no Brasil!')
elif nome in 'Ana Claúdia Jéssica Júlia':
    print('Belo nome feminino!')
else:
    print('Seu nome é bem normal.')
print('Tenha um bom dia, {}!'.format(nome))
```
