```python
#Estrututa composta TUPLA 'São IMUTAVEIS'
lanche = ('Hamburgue', 'Suco', 'Pizza', 'Pudim')
print(lanche)
#Fatiamento
print(lanche[1])
print(lanche[1:3])
print(lanche[:2])
```

```python
#1 OPÇÃO USO DO 'FOR'
lanche = ('Hamburgue', 'Suco', 'Pizza', 'Pudim')
for comida in lanche:
    print(f'Eu vou comer {comida}')
print('Comi pra caramba!')
```

```python
#2 OPÇÃO USO DO 'FOR'
lanche = ('Hamburgue', 'Suco', 'Pizza', 'Pudim', 'Batata Frita')
for cont in range(0, len(lanche)):
    print(f'Eu vou comer {lanche[cont]}')
print('Comi pra caramba!')
```

```python
#3 OPÇÃO USO DO 'FOR'
lanche = ('Hamburgue', 'Suco', 'Pizza', 'Pudim', 'Batata Frita')
for pos, comida in enumerate(lanche):
    print(f'Eu vou comer {comida} na posição {pos}')
```

```python
#USO DE 'SORTED'
lanche = ('Hamburgue', 'Suco', 'Pizza', 'Pudim', 'Batata Frita')
print(sorted(lanche))
```

```python
#PROPRIEDADE 'LEN'
a = (2, 5, 4)
b = (5, 8, 1, 2)
c = b + a
print(len(c))
```

```python
#PROPRIEDADE 'COUNT'
a = (2, 5, 4)
b = (5, 8, 1, 2)
c = b + a
print(c.count(5))#quantas vezes aparece o número 5.
```

```python
#PROPRIEDADE 'INDEX'
a = (2, 5, 4)
b = (5, 8, 1, 2)
c = b + a
print(c)
print(c.index(8))
```
