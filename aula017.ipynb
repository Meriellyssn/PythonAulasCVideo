```python
#Listas são mutaveis
num = [2, 4, 9, 0]
num[2] = 3
print(num)
```

```python
#USANDO 'APPEND' PARA ADICIONAR VALORES.
num = [2, 4, 9, 0]
num.append(7)
print(num)
```

```python
#USANDO 'SORT' PARA ORGANIZAR A LISTA NA ORDEM CRESCENTE.
num = [2, 4, 9, 0, 7]
num.sort()
print(num)
```

```python
#USANDO 'SORT c/ REVERSE' PARA ORGANIZAR A LISTA NA ORDEM DECRESCENTE.
num = [2, 4, 9, 0, 7]
num.sort(reverse=True)
print(num)
```

```python
#USANDO 'LEN' PARA SABER A QUANTIDADE DE ELEMENTOS.
num = [2, 4, 9, 0, 7]
print(f'Essa lista tem {len(num)} elementos.')
```

```python
#USANDO 'INSERT' PARA ADICIONAR ELEMENTOS EM UMA DETERMINADA POSIÇÃO.
num = [2, 4, 9, 3, 7]
num.insert(2, 0)
print(num)
```

```python
#USANDO 'POP()' PARA REMOVER O ULTIMO ELEMENTO.
num = [2, 4, 9, 3, 7]
num.pop()
print(num)
```

```python
#USANDO 'POP()' PARA REMOVER DETERMINADO ELEMENTO.
num = [2, 4, 9, 3, 7]
num.pop(2)
print(num)
```

```python
valores = []
valores.append(5)
valores.append(9)
valores.append(4)
for valor in valores:
    print(f'{valor}...', end='')
```

```python
valores = list()
valores.append(5)
valores.append(9)
valores.append(4)
for chave, valor in enumerate(valores):
    print(f'Na posição {chave} encontrei o valor {valor}!')
print('Cheguei ao final da lista.')
```

```python
#OBS: LIGAÇÃO ENTRE LISTAS (se fizer qualquer alteração em uma das listas a outra tbm muda.
a = [3, 6, 9]
b = a
print(f'Lista A: {a}')
print(f'Lista B: {b}')
```

```python
#FAZENDO COPIA DE UMA VARIAVEL
a = [3, 6, 9]
b = a[:]
b[2] = 8
print(f'Lista A: {a}')
print(f'Lista B: {b}')
```
