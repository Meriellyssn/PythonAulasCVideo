```python
#import math
from math import sqrt

num = int(input('Digite um número: '))
#raiz = math.sqrt(num)
raiz = sqrt(num)

print('A raiz de {} é igual a {}.'.format(num, raiz))#mostra a raiz completa
#print('A raiz de {} aredondada para cima é igual a {}.'.format(num, math.ceil(raiz)))#mostra a raiz aredondada para cima.
print('A raiz de {} é igual a {:.2f}.'.format(num, raiz))
```
