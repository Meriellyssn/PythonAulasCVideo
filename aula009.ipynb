```python
frase = 'Curso em Video Python'
print(frase)
print('-' * 20)

print(frase[3])
print('-' * 20)

print(frase[3:13])
print('-' * 20)

print(frase[:13])
print('-' * 20)

print(frase[13:])
print('-' * 20)

print(frase[1:15:2])
print('-' * 20)

print(frase[1::2])
print('-' * 20)
```
